# League of legends - choice of aram champions

```python
# imports

import requests
from bs4 import BeautifulSoup
import pandas as pd
```

### All champions

```python
URL = "https://www.leagueoflegends.com/en-pl/champions/"
page = requests.get(URL)
soup = BeautifulSoup(page.content, "html.parser")

champs = soup.find_all("span", class_="style__Text-n3ovyt-3")
all_champs = [champ.text for champ in champs]
```

### Aram only champions

Attempt of scraping from [here](https://support-leagueoflegends.riotgames.com/hc/en-us/articles/201752854-Howling-Abyss-ARAM-FAQ) unsuccessfull - browser security check

List copied from [here](https://www.leagueoflegends.com/en-us/news/game-updates/patch-10-7-notes/)

```python
aram_champs = "Aatrox, Ahri, Akali, Amumu, Annie, Ashe, Brand, Braum, Caitlyn, Cho'Gath, Darius, Draven, Ekko, Ezreal, Fiora, Fizz, Garen, Graves, Irelia, Janna, Jarvan IV, Jax, Jhin, Jinx, Karma, Karthus, Katarina, Kayle, Kha'Zix, LeBlanc, Lee Sin, Leona, Lucian, Lulu, Lux, Malphite, Maokai, Master Yi, Miss Fortune, Mordekaiser, Morgana, Nautilus, Nidalee, Pantheon, Pyke, Quinn, Renekton, Riven, Ryze, Sivir, Sona, Soraka, Thresh, Tristana, Tryndamere, Twisted Fate, Twitch, Varus, Vayne, Veigar, Vel'Koz, Vladimir, Wukong, Xayah, Zed"
aram_champs = [champ.strip() for champ in aram_champs.split(',')]
```

### Champs not available in aram

```python
not_in_aram = [champ for champ in all_champs if champ not in aram_champs]
```

### Table with all champions info

```python
url_price = "https://leagueoflegends.fandom.com/wiki/List_of_champions"
table = pd.read_html(url_price, match='Blue Essence')[0]
table
```

```python
table.insert(1, 'Description', '')
table['Description'] = table.loc[:, 'Champion']
table['Champion'] = all_champs
table['Description'] = [a.replace(b, '').strip() for a, b in zip(table['Description'], table['Champion'])]
table
```

### Table with only aram unavailable champions

```python
aram_table = table.loc[table['Champion'].isin(not_in_aram)]
aram_table.to_csv('not_aram.csv', index=False)
```
